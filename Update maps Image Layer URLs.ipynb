```python
from arcgis.gis import *
from IPython.display import display
from getpass import getpass
```

```python
user=input("Ingrese usuario:")
portalpassword = getpass()
gis = GIS("https://mapas.msm.gov.ar/portal", user, portalpassword)
```

```python
user=gis.users.me
user
```

```python
items=user.items(max_items=200)
for item in items:
    if item.type == "Feature Service":
        print(item.title," / ",item.type, " / ", item.url)

    
```

```python
usercontent = user.items(max_items=200)
#folders = user.folders
print ("============================================================================================")
print ("Root folder")
print ("============================================================================================")
for item in usercontent:
    print("Estado original",item.title," / ",item.type, " / ", item.url)
    try:
        nuevaurl=item.url.replace("srvmgis04","mapas",1)
        item.update(item_properties={'url':nuevaurl})
    except Exception as e:
        if hasattr(e, 'message'):
            print(e.message)
        else:
            print(e)
    print("Estado actualizado",item.title," / ",item.type, " / ", item.url)
```
