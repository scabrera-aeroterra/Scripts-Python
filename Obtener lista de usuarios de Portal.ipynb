```python
from arcgis.gis import *
from IPython.display import display
from getpass import getpass
from time import *
```

```python
portalurl=input("Ingrese URL de portal:")
user=input("Ingrese usuario administrador:")
portalpassword = getpass()
```

```python
gis = GIS(portalurl, user, portalpassword)
```

```python
users=gis.users.search("*",max_users=1000)
```

```python
i=0
print ("{:<40s}{:<6s}".format("user.username", "Index"))
print ("{:<40s}{:<6s}".format("-"*14, "-"*5))
for user in users:
    print ("{:<40s}{:>5d}".format(user.username, i))
    i=i+1
        
```
